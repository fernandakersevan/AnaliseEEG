```python
n = 0
for i in range (1, 115):
  if not np.isnan(electrodes1['x'].at[n]):
      plt.plot(electrodes1['x'], electrodes1['y'], 'x')
      plt.text(electrodes1['x'].at[n], electrodes1['y'].at[n], electrodes1['name'].at[n])
  n = n + 1
```
