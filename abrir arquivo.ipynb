```python
sinais = ['KNEE', 'ANKLE', 'HIP', 'Fp1']

for i in range(0, len(sinais)):
    print(sinais[i])
    sinal = data.get_data(sinais[i]).squeeze()
    fs = 512
    t = np.arange(0, len(sinal))/fs
    plt.figure()
    plt.plot(t, sinal)
    plt.xlim(1, 3)
    plt.title(sinais[i])
    plt.show()
```

```python
i = 0
while i < len(sinais):
    print(sinais[i])
    sinal = data.get_data(sinais[i]).squeeze()
    fs = 512
    t = np.arange(0, len(sinal))/fs
    plt.figure()
    plt.plot(t, sinal)
    plt.xlim(1, 3)
    plt.title(sinais[i])
    plt.show()
    i = i + 1
```
