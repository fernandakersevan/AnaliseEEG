# Remoção do artefato ocular

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sinais1=['Fp1']
# Data Acquisition
# Import Data
sinal1 = data.get_data(sinais1).T #Fz|Cz|O2|HEOG
# Data Extraction
eog_HEOG = (-sinal1 - data.get_data('Fp2').T)/2  # Fp1, Fp2, média dos dois
eeg_Fz = data.get_data('Fz').T   # FC1

# RLS ALGORITHM EXECUTION FOR Denoising at Fz is a easy, stable and fast convergence method suitable for online removal of EOG artifacts
############################################################################
# Execution for Fz electrode
# Variable Initialization
sample_no = len(eeg_Fz)  # No of samples/time points
order = 3                # Order of the Adaptive Filter (User Tunable)
sigma = 0.01             # Initializing variable
lambda_ = 0.9999         # Forgetting Factor for RLS Algorithm (User Tunable)
H = np.zeros((order, 1))  # Initial filter Coefficients
R = sigma * np.eye(order, order)  # Initial value for Reference combination

# RLS Algorithm for Adaptive Denoising
correct_Fz = np.zeros(sample_no)  # Initialize the denoised signal
for n in range(sample_no):        # Loop to simulate reality situation
    s = eeg_Fz[n]         # eeg @ Fz at that time point
    if n >= order-1:
        j = 1          # calculation of last "order" reference vector
        r = np.zeros((order, 1))
        for i in range(n, n-order, -1):
            r[j-1, 0] = eog_HEOG[i]
            j = j+1
        # Calculation of Filter Coefficients
        K = np.linalg.inv(R) @ r @ np.linalg.inv(lambda_ + r.T @ np.linalg.inv(R) @ r)
        e = s - (r.T @ H)
        H = H + (K * e)
        R = np.linalg.inv((1/lambda_ * np.linalg.inv(R)) - ((1/lambda_) * K * (r.T) @ np.linalg.inv(R)))
        # Calculation Of RLS algorithm estimated signal
        correct_Fz[n] = s - (r.T @ H)
##################################################################################
# Fz Denoising plot
plt.figure(figsize=(8, 6))
plt.subplot(3, 1, 1)  # HEOG plot
plt.plot(eog_HEOG)
plt.title('Horizontal EOG Artifact')
plt.xlabel('sample number')
plt.ylabel('HEOG mu volts')

plt.subplot(3, 1, 2)  # Raw at Fz plot
plt.plot(eeg_Fz)
plt.title('Raw EEG at Fz')
plt.xlabel('sample number')
plt.ylabel('EEG mu volts')

plt.subplot(3, 1, 3)  # denoised eeg at Fz plot
plt.plot(correct_Fz)
plt.title('Corrected EEG at Fz')
plt.xlabel('sample number')
plt.ylabel('EEG mu volts')

plt.tight_layout()
plt.show()
```

```python
def remover_artefato(EEG, olho):

    sample_no = len(EEG)
    order = 3
    sigma = 0.01
    lambda_ = 0.9999
    H = np.zeros((order, 1))
    R = sigma * np.eye(order, order)

    EEGlimpo = np.zeros(sample_no)
    for n in range(sample_no):
        s = EEG[n]
        if n >= order-1:
            j = 1          # calculation of last "order" reference vector
            r = np.zeros((order, 1))
            for i in range(n, n-order, -1):
                r[j-1, 0] = olho[i]
                j = j+1
            # Calculation of Filter Coefficients
            K = np.linalg.inv(R) @ r @ np.linalg.inv(lambda_ + r.T @ np.linalg.inv(R) @ r)
            e = s - (r.T @ H)
            H = H + (K * e)
            R = np.linalg.inv((1/lambda_ * np.linalg.inv(R)) - ((1/lambda_) * K * (r.T) @ np.linalg.inv(R)))
            # Calculation Of RLS algorithm estimated signal
            EEGlimpo[n] = s - (r.T @ H)


    return EEGlimpo
```
